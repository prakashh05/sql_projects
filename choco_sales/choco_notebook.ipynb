-- Q1 The revenue generated each month in each year

```SQL
SELECT 
    EXTRACT(YEAR FROM date::date) AS year,
    EXTRACT(MONTH FROM date::date) AS month,
    SUM(REPLACE(REPLACE(amount, '$', ''), ',', '')::numeric) AS total_sales
FROM choco_sales
GROUP BY year, month
ORDER BY year, month;

![Result1](screenshots/q1.png)

-- Q2 The revenue generated each month in each year

```SQL
WITH yearly_sales AS (
    SELECT
        EXTRACT(YEAR FROM date::date) AS year,
        SUM(REPLACE(REPLACE(amount, '$', ''), ',', '')::numeric) AS total_sales
    FROM choco_sales
    GROUP BY year
)

SELECT
    year,
    total_sales,
    LAG(total_sales) OVER (ORDER BY year) AS previous_year_sales,
    ROUND(
        (total_sales - LAG(total_sales) OVER (ORDER BY year))
        / LAG(total_sales) OVER (ORDER BY year) * 100,
        2
    ) AS yoy_growth_percent
FROM yearly_sales
ORDER BY year;

![Result2](screenshots/q2.png)
